### Q3. You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

```python
# Creating a histogram for order_hour_of_day column`
hist_ord_hr_of_day = df_ords_prods_custmr['order_hour_of_day'].plot.hist(bins = 70)
```

#### The histogram above shows the frequency of orders places during the day. This shows us that the max orders are placed between 10 am -15 pm.

### Q4. The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
# Creating a bar chart from the "loyalty flag column" for the maeketing team
bar_loyalty_flag = df_ords_prods_custmr['loyalty_flag'].value_counts().plot.bar()
```

### Q5. Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)

```python
np.random.seed(4)
```

```python
#Create a list holding True/False values to test the np.random.rand() <= 0.8
dev = np.random.rand(len(df_ords_prods_custmr)) <= 0.8
```

```python
#Store 80% of the sample in the dataframe big
big = df_ords_prods_custmr[dev]
```

```python
#Store 20% of the sample in the dataframe small
small = df_ords_prods_custmr[~dev]
```

```python
len(df_ords_prods_custmr)
```

```python
len(big) + len(small)
```

```python
# Creating a subset with prices and orders of the day for the line chart
df_prices_ords_hr_of_day = small[['prices', 'order_hour_of_day']]
```

```python
line_prices_ords_hr_of_day = sns.lineplot(data = df_prices_ords_hr_of_day, x = 'order_hour_of_day' , y = 'prices')
```

```python
# Creating a subset with age and number of dependents
df_age_dependents = small[['age', 'dependents']]
```

```python
line_chart_age_dependents = sns.lineplot(data = df_age_dependents, x = 'age' , y = 'dependents')
```

#### From the above line chart we cannot target the audience as the graph is very zigzag and no logical interpretation can be deduced.

```python
scatterplot_age_income = sns.scatterplot(x = 'age' , y = 'income' , data = df_ords_prods_custmr)
```

#### From the above scatterplot we see that as the age increases so does the income level, but we also see that a great majority of this income is concentrated between 200000 and 300000.

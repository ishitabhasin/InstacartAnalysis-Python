# If-Stements with User-Defined functions

```python
#Define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#Checking for the most expensive product in the subset created
df['prices'].max()
```

Q2 : Create 'busiest_days' column with two most and two least busy days

```python
# findind out frequency or orders on different days of the week
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

From the above frequency table we can see that days 0 & 1 (saturday & sunday) are the busiest while days 3 & 4 (Tuesday & Wednesday) are the least busy.

```python
#creating a new column for busiest days
days = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value ==1:
    days.append("Busiest days")
  elif value == 4 or value ==3:
    days.append("Least busy days")
  else:
    days.append("Regularly busy days")
```

```python
days
```

```python
#creating a new column in ords_prods_merge as days
ords_prods_merge['busiest_days'] = days
```

```python
#Checking the frequency 
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

# Busiest hours of day by Period of time 

```python
# findind out frequency of busiest hours of day 
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Since we need to group the 24 hours into 3 categories we see that "Most Orders" (10,11,14,15,13,12,16 &9) "Average Orders" (17,8,18,19,20,7,21 and 22) and "Least Orders" (23,6,0,1,5,2,4 & 3).

```python
#creating a new column for period of time
hours = []

for hour in ords_prods_merge['order_hour_of_day']:
  if hour in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif hour in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
#creating a new column in ords_prods_merge as busiest period of the day
ords_prods_merge['busiest_period_of_day'] = hours
```

```python
# findind out frequency of busiest hours of day 
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
